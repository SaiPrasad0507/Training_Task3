5.Recursion Concepts

Recursion is a technique where a function calls itself.

It is useful for solving problems that can be broken into smaller similar problems.

```python
def factorial(n):
  if n == 1:
    return 1
  return n * factorial(n - 1)
print(factorial(5))
```

```python
def test():
  x = 10
  print(x)
```

Global Variable

A global variable is created outside the function and can be accessed throughout the program.

```python
x = 20
def show():
   print(x)
```

Nonlocal Variable

A nonlocal variable is used in nested functions to modify variables from the outer function.

```python
def outer():
  x = "Python"
  def inner():
    nonlocal x
    x = "Programming"
    inner()
    print(x)
    outer()
```

1. BMI Calculator Function


```python
def bmi_calculator(weight, height):
    bmi = weight / (height ** 2)

    if bmi < 18.5:
        category = "Underweight"
    elif bmi < 25:
        category = "Normal Weight"
    elif bmi < 30:
        category = "Overweight"
    else:
        category = "Obese"

    return bmi, category

weight = float(input("Enter weight in kg: "))
height = float(input("Enter height in meters: "))

bmi, category = bmi_calculator(weight, height)

print("BMI:", round(bmi, 2))
print("Category:", category)
```

2. EMI Calculator

Formula:

 EMI=[PxRx(1+R)^N]/[(1+R)^N]-1




	​


```python
def emi_calculator(principal, annual_rate, years):
    monthly_rate = annual_rate / (12 * 100)
    months = years * 12

    emi = (principal * monthly_rate * (1 + monthly_rate) ** months) / \
          ((1 + monthly_rate) ** months - 1)

    return emi

p = float(input("Enter Loan Amount: "))
r = float(input("Enter Annual Interest Rate: "))
t = int(input("Enter Tenure in Years: "))

emi = emi_calculator(p, r, t)

print("Monthly EMI =", round(emi, 2))
```

3. Recursive Factorial Program

n!=n×(n−1)!

```python
def factorial(n):
    if n == 0 or n == 1:
        return 1
    return n * factorial(n - 1)

num = int(input("Enter a number: "))

print("Factorial =", factorial(num))
```

4. Fibonacci Series Generator

Using Functions

```python
def fibonacci(n):
    a, b = 0, 1

    for i in range(n):
        print(a, end=" ")
        a, b = b, a + b

terms = int(input("Enter number of terms: "))
fibonacci(terms)
```

Fibonacci Using Recursion

```python
def fibonacci_recursive(n):
    if n <= 1:
        return n
    return fibonacci_recursive(n-1) + fibonacci_recursive(n-2)

terms = int(input("Enter terms: "))

for i in range(terms):
    print(fibonacci_recursive(i), end=" ")
```

5. Prime Number Checker

```python
def is_prime(num):

    if num <= 1:
        return False

    for i in range(2, int(num ** 0.5) + 1):
        if num % i == 0:
            return False

    return True

number = int(input("Enter a number: "))

if is_prime(number):
    print(number, "is Prime")
else:
    print(number, "is Not Prime")
```

6. List Operations Functions

Using map()

```python
numbers = [1, 2, 3, 4, 5]

squares = list(map(lambda x: x**2, numbers))

print(squares)
```

Using filter()

```python
numbers = [1, 2, 3, 4, 5, 6]

even = list(filter(lambda x: x % 2 == 0, numbers))

print(even)
```

Using reduce()

```python
from functools import reduce

numbers = [1, 2, 3, 4]

sum_all = reduce(lambda x, y: x + y, numbers)

print(sum_all)
```
